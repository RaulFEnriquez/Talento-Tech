```python
import pandas as pd

value_n_list = ['Andrew', 'Philip', 'Jhon']
print(value_n_list)

value_w_list = [50,        78,     91.5]
value_h_list = [1.76,     1.87,    1.69]

data_dict = {'name': value_n_list,
             'weight': value_w_list,
             'height': value_h_list}

print(data_dict)

df = pd.DataFrame(data_dict)
print(df)
df
```

```python
import pandas as pd

value_n_list = ['Papas', 'Banana', 'Carne', 'Pollo', 'Uvas', 'Arroz', 'Huevos', 'lentejas', 'Frijoles', 'Manzana']
print(value_n_list)

value_w_list = [600, 300, 750, 550, 350, 900, 800, 650, 700, 200]
value_p_list = [15000, 7000, 17000, 16000, 6000, 12000, 13000, 8000, 9000, 3000]

data_dict = {'product': value_n_list,
             'weight': value_w_list,
             'price': value_p_list}

print(data_dict)

df = pd.DataFrame(data_dict)
print(df)
df
```

```python
df['price_per_kilo'] = df['price'] / (df['weight'] / 1000)
df['price_per_kilo'] = df['price_per_kilo'].round(4)
print(df)
df
```
