Creando Grafica de Coseno

```python
import math
import matplotlib.pyplot as plt

lista_eje_x = []
lista_eje_y = []

for i in range(0,720,1):
    angulo=math.radians(i)
    x=angulo
    y=math.cos(x)
    lista_eje_x.append(x)
    lista_eje_y.append(y)

plt.plot(lista_eje_x,lista_eje_y)

plt.xlabel("Angulo (Radianes)")
plt.ylabel("Función Coseno: y=cos(x)")
plt.title("Grafica de la función de Coseno")
plt.show()
plt.savefig("gradica_coseno_x.jpg")
plt.close()
```

Creando grafica de funciones: lienal, cuadratica, cúbica, rais cuadrada

```python
import math as m
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace (0,2,100)
plt.plot(x,x, label="Lineal")
plt.plot(x,x**2, label="Cuadratica")
plt.plot(x,x**3, label= "cúbica")
plt.plot(x,x**(1/2), label="raiz cuadrada")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graficas de funciones: lineal cuadratica, cúbica, raiz cuadrada (Creado por Danilo Enriquez)")
plt.legend()
plt.savefig("graficas_funciones_Danilo.png")
plt.show()
plt.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace (-3,3,100)
plt.plot(x,x, label="Lineal")
plt.plot(x,x**2, label="Cuadratica")
plt.plot(x,x**3, label= "cúbica")
plt.plot(x,np.sqrt(x), label="raiz cuadrada")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graficas de funciones: lineal cuadratica, cúbica, raiz cuadrada (Creado por Raul Enriquez)")
plt.legend()
plt.savefig("graficas_funciones_Raul.png")
plt.show()
plt.close()
```
