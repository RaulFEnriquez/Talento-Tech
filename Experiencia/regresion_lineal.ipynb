```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo ampliados: Año y consumo de energía
data = {'Año': [1990, 1992, 1994, 1996, 1998, 2000, 2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018, 2020, 2022],
        'Consumo_Energia': [1000, 1100, 1250, 1350, 1450, 1500, 1650, 1750, 1900, 2000, 2200, 2350, 2500, 2700, 3000, 3300, 3500]}
df = pd.DataFrame(data)

# Preparar los datos para el modelo
X = df[['Año']] # Variable independiente
y = df['Consumo_Energia'] # Variable dependiente

# Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Predecir los valores de Consumo_Energia usando el modelo entrenado
y_pred = modelo.predict(X)

# Mostrar el gráfico de dispersión
plt.scatter(df['Año'], df['Consumo_Energia'], label='Datos Reales')

# Trazar la línea de regresión
plt.plot(df['Año'], y_pred, color='red', label='Línea de Regresión')

plt.xlabel('Año')
plt.ylabel('Consumo de Energía')
plt.title('Año vs. Consumo de Energía con Línea de Regresión (Datos Ampliados)')
plt.legend() # Mostrar la leyenda para identificar los puntos y la línea
plt.show()
```
